```python
%matplotlib inline
import imageio
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np
import statistics
```

```python
import pandas as pd
```

```python
# identify cold before image paths

target_dir = '/Users/jonathanlifferth/Documents/Health/beard_color/small'

cold_before_paths = []
cold_after_paths = []
for root, dirs, files in os.walk(target_dir):
    for file in files:
        if file.endswith('cb.jpg'):
            cold_before_paths.append(os.path.join(root,file))
        elif file.endswith('ca.jpg'):
            cold_after_paths.append(os.path.join(root,file))

# cold_before_paths = cold_before_paths.sort()
# cold_after_paths = cold_after_paths.sort()

cold_before_paths.sort()
cold_after_paths.sort()

print(cold_before_paths)
print()
print(cold_after_paths)
```

```python
for path in cold_before_paths:
    print(path)
```

```python
# display all images 

for path in cold_before_paths:
    filename = path
    pic = imageio.imread(filename)
    plt.figure(figsize = (5,5))
    plt.imshow(pic)
    
```

```python
# ensure that image parameters match

for path in cold_before_paths:
    filename = path
    pic = imageio.imread(filename)

    print('Type of the image : ' , type(pic)) 
    print('Shape of the image : {}'.format(pic.shape)) 
    print('Image Hight {}'.format(pic.shape[0])) 
    print('Image Width {}'.format(pic.shape[1])) 
    print('Dimension of Image {}'.format(pic.ndim))
```

```python
# view red channel value distributions
for path in cold_before_paths:
    filename = path
    pic = imageio.imread(filename)
    r_channel = np.asarray(pic[ : , : , 0])
    r_channel = r_channel.ravel().tolist()
    plt.figure(figsize = (5,7))
    plt.imshow(pic)
    
    # subsample is used because full image distributions are too slow to visualize
    subsample = r_channel[::1000]
    sns.displot(subsample, binwidth=3)
    plt.show()
 
```

```python
# create list to hold cold before images

cold_before_images = []
for path in cold_before_paths:
    filename = path
    pic = imageio.imread(filename)
    cold_before_images.append(pic)

# images
```

```python
# practice writing a single image to file

# imageio.imwrite('../assets/2022-01-09_11-18cb.png', [images])
```

```python
# view red value summary statistics by image

cold_before_means = []
cold_before_medians = []

for pic in cold_before_images:
    r_channel = np.asarray(pic[ : , : , 0])
    r_channel = r_channel.ravel().tolist()
    mean = np.mean(r_channel)
    cold_before_means.append(mean)
    median = np.median(r_channel)
    cold_before_medians.append(median)

print(cold_before_means)
print(cold_before_medians)
```

```python
for path in cold_after_paths:
    print(path)
```

```python
# create list to hold cold after images

cold_after_images = []
for path in cold_after_paths:
    filename = path
    pic = imageio.imread(filename)
    cold_after_images.append(pic)
```

```python
cold_after_means = []
cold_after_medians = []

for pic in cold_after_images:
    r_channel = np.asarray(pic[ : , : , 0])
    r_channel = r_channel.ravel().tolist()
    mean = np.mean(r_channel)
    cold_after_means.append(mean)
    median = np.median(r_channel)
    cold_after_medians.append(median)
```

```python
cold_dict = {'cb_mean': cold_before_means, 'ca_mean': cold_after_means,
            'cb_median': cold_before_medians, 'ca_median': cold_after_medians}

cold_df = pd.DataFrame(cold_dict)
cold_df
```

```python
cold_df['mean_Δ'] = cold_df['ca_mean'] - cold_df['cb_mean']
cold_df['median_Δ'] = cold_df['ca_median'] - cold_df['cb_median']
```
